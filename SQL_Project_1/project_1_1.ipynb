```python
#
# function to run a select query and return rows in a pandas dataframe
# pandas puts all numeric values from postgres to float
# if it will fit in an integer, change it to integer
#

def my_select_query_pandas(query, rollback_before_flag, rollback_after_flag):
    "function to run a select query and return rows in a pandas dataframe"
    
    if rollback_before_flag:
        connection.rollback()
    
    df = pd.read_sql_query(query, connection)
    
    if rollback_after_flag:
        connection.rollback()
    
    # fix the float columns that really should be integers
    
    for column in df:
    
        if df[column].dtype == "float64":

            fraction_flag = False

            for value in df[column].values:
                
                if not np.isnan(value):
                    if value - math.floor(value) != 0:
                        fraction_flag = True

            if not fraction_flag:
                df[column] = df[column].astype('Int64')
    
    return(df)

# Connect to the Postgres databaseÂ¶
connection = psycopg2.connect(
    user = "postgres",
    password = "ucb",
    host = "postgres",
    port = "5432",
    database = "postgres"
)

# Create a cursor for the connection
cursor = connection.cursor()


# Format float displays in a dataframe
pd.options.display.float_format = '$ {:,.2f}'.format

# Create a one time SQL View between stores and sales table for querries in this notebook

connection.rollback()

query = """

drop view if exists v_join_s_sa;

create view v_join_s_sa
as
select s.store_id,
       s.street as store_street,
       s.city as store_city,
       s.state as store_state,
       s.zip as store_zip,
       s.latitude as store_latitude,
       s.longitude as store_longitude,
       sa.sale_id,
       sa.customer_id,
       sa.sale_date,
       extract(dow from sa.sale_date) as dow,
       to_char(sa.sale_date, 'Day') as day_of_week,
       extract(month from sa.sale_date) as month_number,
       to_char(sa.sale_date, 'Month') as month_name,
       sa.total_amount
from stores s
     join sales as sa
        on s.store_id = sa.store_id
;

"""

cursor.execute(query)

connection.commit()
```

# 1.1.2 Total Sales as a Dollar Amount by Store

Sort by store name in alphabetical order

```python
# Total Sales as a Dollar Amount by Store is a query from the view "v_join_s_sa".
# Python.loc method used to display just store_name and total_sales

rollback_before_flag = True
rollback_after_flag = True

query = """

select store_id, store_city as store_name, sum(total_amount) as total_sales
from v_join_s_sa
group by store_id, store_city
order by store_city

"""

my_select_query_pandas(query, rollback_before_flag, rollback_after_flag).loc[:,['store_name','total_sales']]
```

# 1.1.3 Total Sales as a Dollar Amount by Month

```python
# Total Sales as a Dollar Amount by Month is a query from the view "v_join_s_sa".
# Python.loc method used to display just month and total_sales

rollback_before_flag = True
rollback_after_flag = True

query = """

select month_number as month_index, month_name as month, sum(total_amount) as total_sales
from v_join_s_sa
group by month_number, month_name
order by month_number

"""
my_select_query_pandas(query, rollback_before_flag, rollback_after_flag).loc[:,['month','total_sales']]
```

# Visualization of Monthly Seasonality Trends for AGM

Based on 2020 data, August, October and March are seasonaly the high points of AGM sales and February and September are the low points of annual sales.

```python
# Build data frame for charting

rollback_before_flag = True
rollback_after_flag = True

query = """

select month_number as month_index, month_name as month, sum(total_amount) as total_sales
from v_join_s_sa
group by month_number, month_name

"""
df=my_select_query_pandas(query, rollback_before_flag, rollback_after_flag)
```

```python
# plotting sales

plt.figure(figsize=(15,8))
sns.set_theme(style="darkgrid")
sns.lineplot(x="month", y="total_sales",data=df)
plt.xticks(rotation='horizontal')
plt.title("2020 Total AGM Sales as a Dollar Amount by Month")
plt.xlabel('Month')
plt.ylabel('2020 Annual Sales in $')
plt.show()
```
